```python
# prompt: select rows from employees where "Last Login Time" is before 8am

import pandas as pd
import datetime as dt
employees = pd.read_csv('https://raw.githubusercontent.com/sivabalanb/Data-Analysis-with-Pandas-and-Python/refs/heads/master/employees.csv')
employees['Start Date'] = pd.to_datetime(employees['Start Date'],format="%m/%d/%Y")
employees['Last Login Time'] = pd.to_datetime(employees['Last Login Time'], format="%H:%M %p").dt.time

# Select rows where "Last Login Time" is before 8am
before_8am = employees[employees['Last Login Time'] < dt.time(8, 0)]
print(before_8am)
```

```python
# prompt: select rows from employees where the employee is female and works in marketing

import pandas as pd
import datetime as dt

employees = pd.read_csv('https://raw.githubusercontent.com/sivabalanb/Data-Analysis-with-Pandas-and-Python/refs/heads/master/employees.csv')
employees['Start Date'] = pd.to_datetime(employees['Start Date'],format="%m/%d/%Y")
employees['Last Login Time'] = pd.to_datetime(employees['Last Login Time'], format="%H:%M %p").dt.time

# Select rows where "Gender" is "Female" and "Team" is "Marketing"
female_marketing = employees[(employees["Gender"] == "Female") & (employees["Team"] == "Marketing")]
print(female_marketing)
```
